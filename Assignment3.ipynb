**Question 2**

To derive the conditional distribution $f_{\mathbf{X}\mid\mathbf{Y}}(\mathbf{x} \mid \mathbf{y})$, we use the fundamental definition of conditional density for absolutely continuous distributions:


$$f_{\mathbf{X}\mid\mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) = \frac{f_{\mathbf{X}, \mathbf{Y}}(\mathbf{x}, \mathbf{y})}{f_{\mathbf{Y}}(\mathbf{y})} = \frac{f_{\mathbf{Z}}(\mathbf{z})}{f_{\mathbf{Y}}(\mathbf{y})}$$

The standard way to show this from a joint density perspective is through **block matrix factorization** of the inverse covariance matrix $\boldsymbol{\Sigma}^{-1}$ and the determinant $|\boldsymbol{\Sigma}|$ using **Schur Complements**.

**Factoring the Covariance Matrix ($\boldsymbol{\Sigma}$) and its Determinant:**

The Schur complement of the block $\boldsymbol{\Sigma}_{YY}$ in matrix $\boldsymbol{\Sigma}$ is defined as:


$$\mathbf{S}_{XX} = \boldsymbol{\Sigma}_{X\mid Y} = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$$

We can factor the joint covariance matrix $\boldsymbol{\Sigma}$ into a product of lower triangular, block diagonal, and upper triangular matrices:


$$\boldsymbol{\Sigma} = \begin{bmatrix} \mathbf{I}_n & \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1} \\ \mathbf{0} & \mathbf{I}_m \end{bmatrix} \begin{bmatrix} \boldsymbol{\Sigma}_{X\mid Y} & \mathbf{0} \\ \mathbf{0} & \boldsymbol{\Sigma}_{YY} \end{bmatrix} \begin{bmatrix} \mathbf{I}_n & \mathbf{0} \\ \boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX} & \mathbf{I}_m \end{bmatrix}$$

Taking the determinant of both sides, since the determinant of a block triangular matrix with identity blocks on the diagonal is 1:


$$|\boldsymbol{\Sigma}| = |\boldsymbol{\Sigma}_{X\mid Y}| \cdot |\boldsymbol{\Sigma}_{YY}|$$


**Inverting the Covariance Matrix ($\boldsymbol{\Sigma}^{-1}$):**

Using the matrix factorization above, we can invert $\boldsymbol{\Sigma}$ by inverting each piece of the product in reverse order:


$$\boldsymbol{\Sigma}^{-1} = \begin{bmatrix} \mathbf{I}_n & \mathbf{0} \\ -\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX} & \mathbf{I}_m \end{bmatrix} \begin{bmatrix} \boldsymbol{\Sigma}_{X\mid Y}^{-1} & \mathbf{0} \\ \mathbf{0} & \boldsymbol{\Sigma}_{YY}^{-1} \end{bmatrix} \begin{bmatrix} \mathbf{I}_n & -\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1} \\ \mathbf{0} & \mathbf{I}_m \end{bmatrix}$$

Multiplying these matrices together gives:


$$\boldsymbol{\Sigma}^{-1} = \begin{bmatrix} \boldsymbol{\Sigma}_{X\mid Y}^{-1} & -\boldsymbol{\Sigma}_{X\mid Y}^{-1}\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1} \\ -\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}\boldsymbol{\Sigma}_{X\mid Y}^{-1} & \boldsymbol{\Sigma}_{YY}^{-1} + \boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}\boldsymbol{\Sigma}_{X\mid Y}^{-1}\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1} \end{bmatrix}$$

**Decomposing the Quadratic Form:**

Let $\tilde{\mathbf{x}} = \mathbf{x} - \boldsymbol{\mu}_X$ and $\tilde{\mathbf{y}} = \mathbf{y} - \boldsymbol{\mu}_Y$, so that the joint centered vector is $(\mathbf{z} - \boldsymbol{\mu}) = \begin{bmatrix} \tilde{\mathbf{x}} \\ \tilde{\mathbf{y}} \end{bmatrix}$.

The exponent of the joint density is governed by the quadratic form $Q = (\mathbf{z} - \boldsymbol{\mu})^T \boldsymbol{\Sigma}^{-1} (\mathbf{z} - \boldsymbol{\mu})$. Using our factored representation of $\boldsymbol{\Sigma}^{-1}$, let's see what happens when we look at the rightmost multiplication:


$$\begin{bmatrix} \mathbf{I}_n & -\boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1} \\ \mathbf{0} & \mathbf{I}_m \end{bmatrix} \begin{bmatrix} \tilde{\mathbf{x}} \\ \tilde{\mathbf{y}} \end{bmatrix} = \begin{bmatrix} \tilde{\mathbf{x}} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\tilde{\mathbf{y}} \\ \tilde{\mathbf{y}} \end{bmatrix}$$

Notice that the top term matches exactly the centered conditional mean vector! Let $\mathbf{x} - \boldsymbol{\mu}^* = \tilde{\mathbf{x}} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\tilde{\mathbf{y}}$. Now substitute this back into the full matrix product expression for $Q$:


$$Q = \begin{bmatrix} \mathbf{x} - \boldsymbol{\mu}^* \\ \tilde{\mathbf{y}} \end{bmatrix}^T \begin{bmatrix} \boldsymbol{\Sigma}_{X\mid Y}^{-1} & \mathbf{0} \\ \mathbf{0} & \boldsymbol{\Sigma}_{YY}^{-1} \end{bmatrix} \begin{bmatrix} \mathbf{x} - \boldsymbol{\mu}^* \\ \tilde{\mathbf{y}} \end{bmatrix}$$

Expanding this diagonal block multiplication simplifies the joint exponent perfectly into two distinct independent parts:


$$Q = (\mathbf{x} - \boldsymbol{\mu}^*)^T \boldsymbol{\Sigma}_{X\mid Y}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*) + \tilde{\mathbf{y}}^T \boldsymbol{\Sigma}_{YY}^{-1} \tilde{\mathbf{y}}$$

**Constructing the Division:**

We know that the marginal distribution of a component of a joint multivariate Gaussian is also Gaussian, specifically $\mathbf{Y} \sim \mathcal{N}(\boldsymbol{\mu}_Y, \boldsymbol{\Sigma}_{YY})$:


$$f_{\mathbf{Y}}(\mathbf{y}) = \frac{1}{\sqrt{(2\pi)^m |\boldsymbol{\Sigma}_{YY}|}} \exp\left( -\frac{1}{2} (\mathbf{y} - \boldsymbol{\mu}_Y)^T \boldsymbol{\Sigma}_{YY}^{-1} (\mathbf{y} - \boldsymbol{\mu}_Y) \right)$$

Now, we substitute the decomposed determinant $|\boldsymbol{\Sigma}| = |\boldsymbol{\Sigma}_{X\mid Y}| \cdot |\boldsymbol{\Sigma}_{YY}|$ and the split quadratic form $Q$ into the conditional density equation:

$$f_{\mathbf{X}\mid\mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) = \frac{\frac{1}{\sqrt{(2\pi)^{n+m} |\boldsymbol{\Sigma}_{X\mid Y}||\boldsymbol{\Sigma}_{YY}|}} \exp\left( -\frac{1}{2} (\mathbf{x} - \boldsymbol{\mu}^*)^T \boldsymbol{\Sigma}_{X\mid Y}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*) \right) \exp\left(-\frac{1}{2}(\mathbf{y} - \boldsymbol{\mu}_Y)^T \boldsymbol{\Sigma}_{YY}^{-1} (\mathbf{y} - \boldsymbol{\mu}_Y)\right)}{\frac{1}{\sqrt{(2\pi)^m |\boldsymbol{\Sigma}_{YY}|}} \exp\left( -\frac{1}{2} (\mathbf{y} - \boldsymbol{\mu}_Y)^T \boldsymbol{\Sigma}_{YY}^{-1} (\mathbf{y} - \boldsymbol{\mu}_Y) \right)}$$

Canceling out the $(2\pi)^{m/2}$ powers, the $|\boldsymbol{\Sigma}_{YY}|^{1/2}$ terms, and the entire $\mathbf{y}$-dependent exponential component, we are left strictly with:

$$f_{\mathbf{X}\mid\mathbf{Y}}(\mathbf{x} \mid \mathbf{y}) = \frac{1}{\sqrt{(2\pi)^n |\boldsymbol{\Sigma}_{X\mid Y}|}} \exp\left( -\frac{1}{2} (\mathbf{x} - \boldsymbol{\mu}^*)^T \boldsymbol{\Sigma}_{X\mid Y}^{-1} (\mathbf{x} - \boldsymbol{\mu}^*) \right)$$

Where:

* $\boldsymbol{\Sigma}_{X\mid Y} = \mathbf{S}_{XX} = \boldsymbol{\Sigma}_{XX} - \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}\boldsymbol{\Sigma}_{YX}$
* $\boldsymbol{\mu}^* = \boldsymbol{\mu}_X + \boldsymbol{\Sigma}_{XY}\boldsymbol{\Sigma}_{YY}^{-1}(\mathbf{y} - \boldsymbol{\mu}_Y)$

This proves that the conditional distribution is itself a multivariate normal distribution: $\mathbf{X} \mid \mathbf{Y}=\mathbf{y} \sim \mathcal{N}(\boldsymbol{\mu}^*, \boldsymbol{\Sigma}_{X\mid Y})$.

1.Joint DistributionSince $X$ and $Q$ are independent normal random variables, any linear combination of them is also normally distributed. Because $Y = X + Q$ is a linear combination, the vector $(X, Y)^T$ follows a joint multivariate normal distribution.

Marginal Distribution of $Y$: Since $X \sim \mathscr{N}(0, 1)$ and $Q \sim \mathscr{N}(0, 4)$ are independent:$$\mu_Y = \mathbb{E}[Y] = \mathbb{E}[X + Q] = \mathbb{E}[X] + \mathbb{E}[Q] = 0 + 0 = 0$$$$\text{Var}(Y) = \text{Var}(X + Q) = \text{Var}(X) + \text{Var}(Q) = 1 + 4 = 5$$Thus, $Y \sim \mathscr{N}(0, 5)$.

Covariance $\text{Cov}(X, Y)$:Using the linearity of covariance and the independence of $X$ and $Q$ ($\text{Cov}(X, Q) = 0$):$$\text{Cov}(X, Y) = \text{Cov}(X, X + Q) = \text{Cov}(X, X) + \text{Cov}(X, Q) = \text{Var}(X) + 0 = 1$$The joint distribution is therefore:$$\begin{bmatrix} X \\ Y \end{bmatrix} \sim \mathscr{N}\left( \begin{bmatrix} 0 \\ 0 \end{bmatrix}, \begin{bmatrix} 1 & 1 \\ 1 & 5 \end{bmatrix} \right)$$

2.Information GeometryFor a random variable to be $\sigma(Y)$-measurable, its value must be uniquely determined by knowing $Y$ (by the Doob-Dynkin Lemma, $X$ would have to be a deterministic function of $Y$).Mathematically, the pre-image of an event like $\{X > 2\}$ cannot be represented as an element of the sub-$\sigma$-algebra $\sigma(Y)$. This is because there exist infinite pairs of $(X, Q)$ that produce the exact same value of $Y$.

Since multiple states of reality map to the same observation, $X$ is not measurable with respect to our information pool $\sigma(Y)$.

3. Conditional Expectation

Using the formula for conditional expectation of a bivariate normal distribution:

$$\mathbb{E}[X \mid Y = y] = \mu_X + \frac{\text{Cov}(X,Y)}{\text{Var}(Y)}(y - \mu_Y)$$
Substituting our calculated parameters:

$$\mathbb{E}[X \mid Y = y] = 0 + \frac{1}{5}(y - 0) = \frac{1}{5}y$$
As a random variable, we write:

$$\widehat{X} = \mathbb{E}[X \mid Y] = \frac{1}{5}Y$$Signal-to-Total-Variance Ratio:

$$\mathbb{E}[X \mid Y] = \left(\frac{\sigma_X^2}{\sigma_X^2 + \sigma_Q^2}\right) Y = \left(\frac{1}{1 + 4}\right)Y = \frac{1}{5}Y$$

The conditional expectation acts as a scaling filter where the weight applied to our observation is exactly the proportion of total variance accounted for by the true signal.

4.Verification of the Residual

Let $Z = X - \widehat{X} = X - \frac{1}{5}Y$.

Proof that $\mathbb{E}[Z] = 0$:

$$\mathbb{E}[Z] = \mathbb{E}\left[X - \frac{1}{5}Y\right] = \mathbb{E}[X] - \frac{1}{5}\mathbb{E}[Y] = 0 - \frac{1}{5}(0) = 0$$
Verification of Orthogonality ($\text{Cov}(Z, Y) = 0$):

$$\text{Cov}(Z, Y) = \text{Cov}\left(X - \frac{1}{5}Y, Y\right) = \text{Cov}(X, Y) - \frac{1}{5}\text{Cov}(Y, Y)$$

$$\text{Cov}(Z, Y) = \text{Cov}(X, Y) - \frac{1}{5}\text{Var}(Y) = 1 - \frac{1}{5}(5) = 1 - 1 = 0$$

Because $(Z, Y)$ are jointly Gaussian and their covariance is $0$, the estimation error $Z$ is completely independent of our observation $Y$. No additional linear information can be extracted from $Y$ to improve our guess.

5.Information Theory: Differential Entropy

The differential entropy of a continuous Gaussian random variable $W \sim \mathscr{N}(\mu, \sigma^2)$ is $h(W) = \frac{1}{2}\ln(2\pi e \sigma^2)$.

Entropy $h(X)$:

$$h(X) = \frac{1}{2}\ln(2\pi e \cdot 1) = \frac{1}{2}\ln(2\pi e) \approx 1.4189 \text{ nats}$$

Conditional Entropy $h(X \mid Y)$:
First, find the conditional variance $\sigma_{X|Y}^2$:

$$\sigma_{X|Y}^2 = \sigma_X^2 - \frac{\text{Cov}(X,Y)^2}{\text{Var}(Y)} = 1 - \frac{1^2}{5} = 1 - 0.20 = 0.80$$

Since the conditional variance is independent of the specific value of $y$, the conditional entropy is:

$$h(X \mid Y) = \frac{1}{2}\ln(2\pi e \cdot \sigma_{X|Y}^2) = \frac{1}{2}\ln(2\pi e \cdot 0.80) \approx 1.3073 \text{ nats}$$

---

6. Mutual Information

Calculate $I(X; Y)$:

$$I(X; Y) = h(X) - h(X \mid Y) = \frac{1}{2}\ln(2\pi e) - \frac{1}{2}\ln(2\pi e \cdot 0.80)$$

$$I(X; Y) = \frac{1}{2}\ln\left(\frac{1}{0.80}\right) = \frac{1}{2}\ln(1.25) \approx 0.1116 \text{ nats}$$

Percentage of "Uncertainty" Removed:
In terms of variance reduction (the standard metric for continuous spaces):

$$\text{Uncertainty Removed} = \frac{\text{Var}(X) - \text{Var}(X \mid Y)}{\text{Var}(X)} = \frac{1 - 0.80}{1} = 20\%$$

If Noise Variance Increased to $\sigma_Q^2 = 100$:The conditional variance would become $\sigma_{X|Y}^2 = 1 - \frac{1}{101} = \frac{100}{101} \approx 0.9901$.

The mutual information would drop to:

$$I(X;Y) = \frac{1}{2}\ln\left(\frac{1}{0.9901}\right) = \frac{1}{2}\ln(1.01) \approx 0.005 \text{ nats}$$

The percentage of uncertainty removed would plummet to $\approx 0.99\%$. The sensor reading becomes virtually useless because it is drowned out by mechanical noise.

7.Diagnostic Interpretation

The manager's suggestion to use $X = Y$ is a **Naive Estimator**.
* If we use $X = Y$, our error variance would be $E[(X-Y)^2] = E[Q^2] = 4$.
* By using $E[X|Y] = 0.2Y$, our error variance is only $0.8$.

**Shrinkage:** The factor $0.2$ is called a shrinkage estimator. It tells us that when a sensor is noisy, the "best" guess is to be conservative and pull the reading back toward the historical average (zero). If the power spikes, we don't immediately assume the tool is melting; we assume most of that spike is just the material being tough.
